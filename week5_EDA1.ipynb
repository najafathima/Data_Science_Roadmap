### IQR METHOD

##### The IQR method for outlier detection is a simple, classic technique based on dividing your data into four parts (called quartiles) and looking for values that are unusually far outside the “normal range.”

#### Steps
- Calculating the first quartile (Q1): The value below which 25% of data fall (25th percentile).

- Calculating the third quartile (Q3): The value below which 75% of data fall (75th percentile).

- IQR = Q3 - Q1

##### 1. Get 25th and 75th percentiles (these are Q1 and Q3):

```python
q1=np.percentile(df['ads_listened_per_week'],25)
q3=np.percentile(df['ads_listened_per_week'],75)
```

```python
print(q1,q3)
```

```python
df['ads_listened_per_week'].min()
```

```python
df['ads_listened_per_week'].max()
```

##### 2. Compute IQR:

```python
iqr=q3-q1
iqr
```

##### 3. Calculate fences using percentiles:

```python
lower_bound=q1-(1.5*iqr)
upper_bound=q3+(1.5*iqr)
print(lower_bound,upper_bound)
```

##### 4. Detect outliers using these percentile-based boundaries

```python
outlier=[]
for i in df['ads_listened_per_week']:
    if i<lower_bound or i>upper_bound:
        outlier.append(i)
len(outlier)
```

```python
df['ads_listened_per_week']=df['ads_listened_per_week'].clip(lower=lower_bound,upper=upper_bound)
```

```python


# Create a normal distribution, mean=50, std=10, with 1000 values
np.random.seed(42)
arr = np.random.normal(loc=50, scale=10, size=1000)

# Add some high outliers
arr = np.append(arr, [100, 120, 130, 140])

print(arr)

```

```python
sns.histplot(arr,kde=True)
```

```python
outlier1=[]
for i in arr:
    z=(i-arr.mean())/arr.std()
    if abs(z)>3:
        outlier1.append(i)
len(outlier1)
```

```python
sns.boxplot(arr)
```

```python
outlier1
```

```python
mask=np.isin(arr,outlier1)
indices=np.where(mask)[0]
indices
```

```python
arr=np.delete(arr,indices)
arr
```

```python
sns.histplot(arr,kde=True)
```

```python
sns.boxplot(arr)
```

```python
outlier2=[]
for i in arr:
    z=(i-arr.mean())/arr.std()
    if abs(z)>2:
        outlier2.append(i)
outlier2
```

```python
mask=np.isin(arr,outlier2)
indices1=np.where(mask)[0]
len(indices1)
```

```python
arr=np.delete(arr,indices1)
arr
```
