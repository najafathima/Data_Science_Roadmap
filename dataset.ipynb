```python
import numpy as np
import pandas as pd

# Load your dataset (Modify file path if needed)
df = pd.read_csv("Sample - Superstore.csv",encoding='latin-1')

# ----- ADD MISSING VALUES -----

# Randomly select 2% of rows for missing values
np.random.seed(42)  # for reproducibility

rows = df.sample(frac=0.02).index  # 2% rows

# Add missing values in numeric columns
df.loc[rows, 'Sales'] = np.nan
df.loc[rows, 'Profit'] = np.nan
df.loc[rows, 'Discount'] = np.nan

# Add missing values in categorical columns
df.loc[rows, 'City'] = np.nan
df.loc[rows, 'Category'] = np.nan
```

```python
# ----- ADD OUTLIERS -----

# Add *high* outliers in Sales
outlier_rows = df.sample(20).index
df.loc[outlier_rows, 'Sales'] = df['Sales'].mean() * 25  # 25x higher

# Add *negative* outliers in Profit (huge loss)
loss_rows = df.sample(20).index
df.loc[loss_rows, 'Profit'] = df['Profit'].mean() * -20

# Add quantity outliers
qty_rows = df.sample(15).index
df.loc[qty_rows, 'Quantity'] = df['Quantity'].max() * 10  # 10x higher
```

```python
print(df.isnull().sum())
df.describe()
```

```python
df.to_csv('Superstore.csv')
```
