### Training and Testing data 

```python
import pandas as pd

data = {
    "Mileage": [69000, 35000, 57000, 22500, 46000, 59000, 54000, 72000, 91000, 67000,
                83000, 79000, 58780, 82450, 25480, 28000, 69000, 87600, 52000],

    "Age(yrs)": [6, 3, 5, 2, 3, 5, 5, 6, 8, 6,
                 7, 7, 4, 7, 2, 3, 5, 8, 5],

    "Sell Price($)": [18000, 34000, 26100, 46000, 31500, 26750, 32000, 19300, 12000, 22000,
                      18700, 19500, 27500, 22000, 35500, 35000, 19700, 12800, 28200]
}

df = pd.DataFrame(data)
print(df)
```

```python
# to figure out the relationship between the dependant and independant variables
# Mileage and age -> independant variables
# Sell price -> dependant variables
import matplotlib.pyplot as plt
plt.scatter(df['Mileage'],df['Sell Price($)'])
```

```python
plt.scatter(df['Age(yrs)'],df['Sell Price($)'])
```

```python
x=df[['Mileage','Age(yrs)']]
y=df['Sell Price($)']
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2)
```

```python
len(x_train)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model= LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```
